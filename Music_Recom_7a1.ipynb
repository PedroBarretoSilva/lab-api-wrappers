# Designing a function to get Playlists from Spotify

```python
def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)         # Python function developed byh Spotify
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)                  # Python funtion developed byh Spoticfy
        tracks = tracks + results['items']
        sleep(randint(1,3))                      # scraping in a respectful manner
    return tracks
```
